```python
pd.read_sql_query("""
    SELECT
         EXTRACT(YEAR FROM uis_registration_date) reg_year
        ,case when address is NULL then 0 else 1 end address
        ,case
            when LENGTH(inn) = 10 then 'ul'
            when LENGTH(inn) = 12 then 'ip'
            else 'err'
         end
         kind
        ,count(1) count
    FROM STAGE.ERUZ_MAIN
    where
         EXTRACT(YEAR FROM uis_registration_date) in (2019, 2020, 2021, 2022)
    group by
         EXTRACT(YEAR FROM uis_registration_date)
        ,case when address is NULL then 0 else 1 end
        ,case
            when LENGTH(inn) = 10 then 'ul'
            when LENGTH(inn) = 12 then 'ip'
            else 'err'
         end
    order by
         EXTRACT(YEAR FROM uis_registration_date)
        ,case
            when LENGTH(inn) = 10 then 'ul'
            when LENGTH(inn) = 12 then 'ip'
            else 'err'
         end
    """,
    dwh_connection)
```

```python
pd.read_sql_query("""
    SELECT
         SUBSTR(start_reg_date, 0, 4) reg_year
        ,case when kladr_code is NULL then 0 else 1 end kladr_code
        ,case
            when LENGTH(inn) = 10 then 'ul'
            when LENGTH(inn) = 12 then 'ip'
            else 'err'
         end
         kind
        ,count(1) count
    FROM CORE.ESB_ERUZ_CONTRACTOR
    where
         SUBSTR(start_reg_date, 0, 4) in ('2019', '2020', '2021', '2022')
    group by
         SUBSTR(start_reg_date, 0, 4)
        ,case when kladr_code is NULL then 0 else 1 end
        ,case
            when LENGTH(inn) = 10 then 'ul'
            when LENGTH(inn) = 12 then 'ip'
            else 'err'
         end
    order by
         SUBSTR(start_reg_date, 0, 4)
        ,case
            when LENGTH(inn) = 10 then 'ul'
            when LENGTH(inn) = 12 then 'ip'
            else 'err'
         end
    """,
    dwh_connection)
```

```python
pd.read_sql_query("""
    SELECT
        case
            when LENGTH(inn) = 10 then 'ul'
            when LENGTH(inn) = 12 then 'ip'
            else 'err'
        end
        kind
        , SUBSTR(inn, 0, 2) region
        , count(1) count
    FROM STAGE.ERUZ_MAIN
    where
        SUBSTR(inn, 0, 2) = '66'
    group by
        case
            when LENGTH(inn) = 10 then 'ul'
            when LENGTH(inn) = 12 then 'ip'
            else 'err'
        end
        , SUBSTR(inn, 0, 2)
    """,
    dwh_connection)
```

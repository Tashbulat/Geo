```python
def kladr_to_coodr(kladr):
    response = dadata.find_by_id("address", kladr)
    address = response[0]['unrestricted_value']
    latitude = response[0]['data']['geo_lat']
    longitude = response[0]['data']['geo_lon']
    return address, latitude, longitude
```

```python
path = '/home/tochka/Documents/Other/Work/Geo/EGRUL/kladr'

for reg in regions:
    
    print('-----------------')
    print(reg)
    
    kladr_files = os.listdir(path)
    
    if f'df_kladr_{reg}_upd.csv.gz' in kladr_files:
        print('already exists')
        continue
            
    elif f'df_kladr_{reg}.csv.gz' not in kladr_files:
        print('no kladr file')
        continue
    
    else:
        df_kladr = pd.read_csv(
            f'{path}/df_kladr_{reg}.csv.gz',
            compression='gzip',
            dtype={'kladr_11':str}
        )
    
        df_kladr['kladr'] = df_kladr['kladr_11'] + '00'
        
        kladr_list = df_kladr['kladr']
        
        addresses = []
        latitudes = []
        longitudes = []

        for kladr in tqdm(kladr_list):
            try:
                address, latitude, longitude = kladr_to_coodr(kladr)
                addresses.append(address)
                latitudes.append(latitude)
                longitudes.append(longitude)
                
            except Exception as e:
                addresses.append(None)
                latitudes.append(None)
                longitudes.append(None)
            
        df_kladr['address'] = addresses
        df_kladr['latitude'] = latitudes
        df_kladr['longitude'] = longitudes
        
        print(df_kladr.head())
        
        df_kladr.to_csv(
            f'{path}/df_kladr_{reg}_upd.csv.gz',
            sep=';',
            index=False,
            compression='gzip'
        )
    
```

# FIAS

```python
def fias_to_coodr(fias):
    response = dadata.find_by_id("address", str(fias))
    address = response[0]['unrestricted_value']
    latitude = response[0]['data']['geo_lat']
    longitude = response[0]['data']['geo_lon']
    return address, latitude, longitude
```

```python
path = '/home/tochka/Documents/Other/Work/Geo/EGRUL/fias'

for reg in regions:
    
    print('-----------------')
    print(reg)
    
    kladr_files = os.listdir(path)
    
    if f'df_fias_{reg}_upd.csv.gz' in kladr_files:
        print('already exists')
        continue
            
    elif f'df_fias_{reg}.csv.gz' not in kladr_files:
        print('no fias file')
        continue
    
    else:
        df_fias = pd.read_csv(
            f'{path}/df_fias_{reg}.csv.gz',
            compression='gzip',
            dtype={'kladr_11':str}
        )
        
        fias_list = df_fias['fias_code']
        
        addresses = []
        latitudes = []
        longitudes = []

        for fias in tqdm(fias_list):
            try:
                address, latitude, longitude = fias_to_coodr(fias)
                addresses.append(address)
                latitudes.append(latitude)
                longitudes.append(longitude)
                
            except Exception as e:
                addresses.append(None)
                latitudes.append(None)
                longitudes.append(None)
            
        df_fias['address'] = addresses
        df_fias['latitude'] = latitudes
        df_fias['longitude'] = longitudes
        
        print(df_fias.head())
        
        df_fias.to_csv(
            f'{path}/df_fias_{reg}_upd.csv.gz',
            sep=';',
            index=False,
            compression='gzip'
        )
    
```
